### 5a) Calculate the average "Engine HP" in the dataset.

```python
print('Average Engine HP:',df['Engine HP'].mean())
```

### 5b) Use the fillna method and to fill the missing values in "Engine HP" with the mean value from the previous step.

```python
average_engine_hp = df['Engine HP'].mean()
print(average_engine_hp)
```

### 5c) Now, calcualte the average of "Engine HP" again.

```python
df['Engine HP'] = df['Engine HP'].fillna(average_engine_hp)
```

### 5d) Has it changed?

```python
print('Average Engine HP Change after filling average in for NA values:',df['Engine HP'].mean() != average_engine_hp)
```

### 6a) Select all the "Rolls-Royce" cars from the dataset.

```python
rolls_royce_df = df[df['Make'] == 'Rolls-Royce']
rolls_royce_df
```

### 6b) Select only columns "Engine HP", "Engine Cylinders", "highway MPG".

```python
engine_hp_cyl_hwyMPG = rolls_royce_df[['Engine HP','Engine Cylinders','highway MPG']]
```

### 6c) Now drop all duplicated rows using drop_duplicates method (you should get a dataframe with 7 rows).

```python
engine_hp_cyl_hwyMPG= engine_hp_cyl_hwyMPG.drop_duplicates()
```

### 6d) Get the underlying NumPy array. Let's call it X.

```python
X = engine_hp_cyl_hwyMPG.values
X
```

### 6e) Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.

```python
XTX = X.T.dot(X)
XTX
```

### 6f) Invert XTX.

```python
XTX_inv = np.linalg.inv(XTX)
```

### 6g) What's the sum of all the elements of the result?

```python
XTX_inv.sum()
```

### 7a) Create an array y with values [1000, 1100, 900, 1200, 1000, 850, 1300].

```python
y = np.array([1000, 1100, 900, 1200, 1000, 850, 1300])
```

### 7b) Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.


```python
w = XTX_inv.dot(X.T).dot(y)
w
```
